```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
## We will set a random seed and defined, a mean= 0 and variance= 1.With the numpy random we can create a normal distribution.
## And with this plot, we can plot the data

np.random.seed(0)
mu, sigma = 0,1

data = np.random.normal(mu, sigma, 1000)
```

```
p = sns.distplot(data)
```

```
p = sns.distplot(data, color = '#F15B2A')
```

```
p = sns.distplot(data, color = '#F15B2A', hist_kws = dict(edgecolor = '#F15B2A'))
```

```
p = sns.distplot(data, color = '#F15B2A', hist_kws = dict(edgecolor = '#F15B2A', linewidth = 2))
```

```
p = sns.distplot(data, color = '#F15B2A', hist_kws = dict(edgecolor = '#F15B2A', linewidth = 2), bins = 50)
```

```
p = sns.distplot(data,  
                 color = '#F15B2A',
                 hist_kws = dict(edgecolor = '#F15B2A', linewidth = 2),
                 bins = 50,
                 kde=True,
                 kde_kws = dict(shade=True)
                 )
```

```
p = sns.kdeplot(data, color = '#F15B2A')
```

```
p = sns.kdeplot(data,
                color = '#F15B2A',
                shade= True)
```

```
p = sns.boxplot(data)
```

```
p = sns.boxplot(data, palette = ['#F15B2A'])
```

```
p = sns.boxplot(data,
                palette=["#F15B2A"])

for e in p.artists:
  r, g, b, a = e.get_facecolor()
  e.set_facecolor((r, g, b, .0))
```

```
p = sns.boxplot(data,
                palette=["#F15B2A"])

for e in p.artists:
  r, g, b, a = e.get_facecolor()
  e.set_facecolor((r, g, b, .0))

for i, artist in enumerate(p.artists):
  artist.set_edgecolor("#F15B2A")
```

```
p = sns.boxplot(data,
                palette=["#F15B2A"])

for e in p.artists:
  r, g, b, a = e.get_facecolor()
  e.set_facecolor((r, g, b, .1))

for i, artist in enumerate(p.artists):
  artist.set_edgecolor("#F15B2A")
```

```
p = sns.boxplot(data,
                palette=["#F15B2A"])

for e in p.artists:
  r, g, b, a = e.get_facecolor()
  e.set_facecolor((r, g, b, .3))

for i, artist in enumerate(p.artists):
  artist.set_edgecolor("#F15B2A")

for j in range(0, 6):
  line = p.lines[j]
  line.set_color('#F15B2A')
  line.set_mfc('#F15B2A')
  line.set_mec('#F15B2A')
```

```
p =sns.violinplot(data)
```

```
p =sns.violinplot(data, palette = ['#F15B2A'])
```

```
# Using matplotlib to change elements properties

parts = plt.violinplot(data, vert = False, showmeans = False, showmedians = True);

for p in parts['bodies']:
  p.set_facecolor('#F15B2A')
  p.set_edgecolor('#F15B2A')
  p.set_alpha(0.3)
```

```
tips = sns.load_dataset('tips')

p = sns.stripplot(x=tips['total_bill'],
                  color = '#F15B2A',
                  jitter = 0)
```

```
tips = sns.load_dataset('tips')

p = sns.swarmplot(x=tips['total_bill'],
                  color = '#F15B2A')
```
